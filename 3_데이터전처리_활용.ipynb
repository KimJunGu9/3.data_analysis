* [민간 아파트 가격동향](https://bit.ly/ds-house-price)

```python
import pandas as pd
```

```python
df = pd.read_csv('https://bit.ly/ds-house-price')
```

```python
df
```

```python
df.info()
```

```python
# 컬럼명 바꾸기  /   rename(columns = {' ~~~ ' : ' ~~~ '})

df = df.rename(columns={'분양가격(㎡)' : '분양가격'})
```

```python
df
```

```python
# 통계값 확인하기

df.describe()
```

```python
# 분양가격을 int로 변환

# df['분양가격'].astype(int)  # ValueError: invalid literal for int() with base 10: '  '
```

```python
# '  ' 공백이 존재하는 데이터가 있는지 확인

df.loc[df['분양가격'] == '  ']
```

```python
# strip()를 사용하여 공백 ('  ')이 있는 데이터의 공백을 삭제하기

df['분양가격'] = df['분양가격'].str.strip()
```

```python
# 공백 ('  ')이 잘 삭제됐는지 확인하기

df.loc[df['분양가격'] == '  ']
```

```python
# df['분양가격'].astype(int)  # ValueError: invalid literal for int() with base 10: ''
```

```python
# 공백 ('') 이 잘 삭제됐는지 확인하기

df.loc[df['분양가격'] == '']
```

```python
# 데이터가 없는 경우는 0을 채워줌

df.loc[df['분양가격'] == '', '분양가격'] = 0
```

```python
# 0이 잘 들어갔는지 확인하기

df.loc[df['분양가격'] == '']
```

```python
# df['분양가격'].astype(int)  # ValueError: cannot convert float NaN to integer
```

```python
# NaN을 0으로 변환

df['분양가격'] = df['분양가격'].fillna(0)
```

```python
# df['분양가격'].astype(int)  # ValueError: invalid literal for int() with base 10: '6,657'
```

```python
# 모든 자료에서 콤마를 제거

df['분양가격'] = df['분양가격'].str.replace(',','')
```

```python
# df['분양가격'].astype(int)  # ValueError: cannot convert float NaN to integer
```

```python
# NaN를 0으로 채우기

df['분양가격'] = df['분양가격'].fillna(0)
```

```python
# df['분양가격'].astype(int)  # ValueError: invalid literal for int() with base 10: '-'
```

```python
# '-'을 제거

df['분양가격'] = df['분양가격'].str.replace('-','')
```

```python
# df['분양가격'].astype(int)  # ValueError: cannot convert float NaN to integer
```

```python
# NaN를 0으로 채우기

df['분양가격'] = df['분양가격'].fillna(0)
```

```python
# df['분양가격'].astype(int) # ValueError: invalid literal for int() with base 10: ''
```

```python
# 데이터가 없는경우('') 0으로 바꾸기

df.loc[df['분양가격'] == '', '분양가격'] = 0
```

```python
df['분양가격'].astype(int)
```

```python
df['분양가격'] = df['분양가격'].astype(int)
```

```python
df.info()
```

```python
df['규모구분']
```

```python
# 규모구분 column에 불필요한 '전용면적' 글자 삭제

df['규모구분'] = df['규모구분'].str.replace('전용면적', '')
```

```python
df['규모구분'].value_counts()
```

```python
# 지역명별로 평균 분양가격을 확인

df.groupby('지역명')['분양가격'].mean()
```

```python
# 지역별 최고 비싼 분양가

df.groupby('지역명')['분양가격'].max()
```

```python
# 연도별 분양가격의 평균

df.groupby('연도')['분양가격'].mean()
```

```python
# 작업파일 저장

df.to_csv('ds_house_price_clean.csv', index=False)
```
