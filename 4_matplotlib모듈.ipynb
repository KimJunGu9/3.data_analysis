```python
plt.plot(np.arange(10), np.arange(10)*2)
plt.plot(np.arange(10), np.arange(10)**2)
plt.plot(np.arange(10), np.log(np.arange(10)))

plt.title('범례 설정', fontsize=30)

plt.xlabel('X', fontsize=20)
plt.ylabel('Y', fontsize=20, rotation=0)

plt.xticks(rotation=90)  # ticks : 눈금 # 90도
plt.yticks(rotation=30)  # 30도

# plt.legend(['10*2', '10**2', 'log'], fontsize=15, loc='upper right')
plt.legend(['10*2', '10**2', 'log'], fontsize=15, loc=(0.5, 0.5), ncol=3)

plt.xlim(0, 10)  # x축 범위
plt.ylim(0.5, 20)

plt.show()
```

# **3. 스타일 세부 설정**

[스타일 세부 설정](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html)

```python
plt.plot(np.arange(10), np.arange(10)*2, marker='o', markersize=5, linestyle='', color='b', alpha=0.3)  # alpha : 투명도
plt.plot(np.arange(10), np.arange(10)**2, marker='v', markersize=10, linestyle='--', color='c', alpha=0.5)
plt.plot(np.arange(10), np.log(np.arange(10)), marker='*', markersize=15, linestyle='-.', color='r', alpha=0.7)

plt.title('마커 설정', fontsize=30)

plt.xlabel('X', fontsize=20)
plt.xlabel('Y', fontsize=20, rotation=0)

plt.xticks(rotation=90)
plt.xticks(rotation=30)

plt.grid() # 도화지가 모눈종이처럼 됨

plt.show()
```

```python
x = np.random.rand(50) # 0 ~ 1사이의 소수
y = np.random.rand(50)

colors = np.arange(50)
area = x * y * 1000

print(x)
print(y)
print(colors)
print(area)
```

```python
plt.scatter(x, y, s=area, c=colors)
plt.show()
```

```python
plt.figure(figsize=(12, 6))

plt.subplot(131)
plt.scatter(x, y, s=area, cmap='blue', alpha=0.1)
plt.title('alpha=0.1')


plt.subplot(132)
plt.scatter(x, y, s=area, cmap='blue', alpha=0.5)
plt.title('alpha=0.5')



plt.subplot(133)
plt.scatter(x, y, s=area, cmap='blue', alpha=1.0)
plt.title('alpha=1.0')

plt.show()
```

```python
x = ['Math', 'Programming', 'Data science', 'art', 'English', 'korea']
y = [67, 80, 50, 80, 10, 90]

plt.figure(figsize=(6, 3))
plt.bar(x, y, align='edge', alpha=0.6, color='red')   # align : 방향, edge로 하면 왼쪽아래끄트머리로 맞춰짐
plt.xticks(x)
plt.ylabel('Number of Students')
plt.title('Subjeects')
plt.show()
```

```python
x = ['Math', 'Programming', 'Data cience', 'art', 'English', 'korea']
y = [67, 80, 50, 80, 10, 90]

plt.figure(figsize=(6, 3))
# plt.bar(x, y, align='edge', alpha=0.6, color='red')
plt.barh(x, y, align='center', alpha=0.7, color='pink')
plt.yticks(x)
plt.xlabel('Number of Students')
plt.title('Subjeects')
plt.show()
```

```python
x = np.arange(0, 10, 0.1)
y = 1 + np.sin(x)

plt.plot(x, y)

plt.xlabel('x value', fontsize=15)
plt.ylabel('y value', fontsize=15)
plt.title('sin graph', fontsize=30)

plt.show()
```
